```python
# Gráfico de barras para ventas promedio en días festivos y no festivos
festivo_vs_no_festivo = df.groupby('Festivo')['Ventas'].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=festivo_vs_no_festivo.index, y=festivo_vs_no_festivo.values, palette='viridis')
plt.title('Promedio de Ventas: Festivo vs No Festivo')
plt.xlabel('Festivo (0 = No, 1 = Sí)')
plt.ylabel('Promedio de Ventas')
plt.grid(True)
plt.show();
```

```python
# Gráfico de barras para ventas promedio con y sin promociones
promociones_vs_no_promociones = df.groupby('Promociones')['Ventas'].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=promociones_vs_no_promociones.index, y=promociones_vs_no_promociones.values, palette='coolwarm')
plt.title('Promedio de Ventas: Con y Sin Promociones')
plt.xlabel('Promociones (0 = No, 1 = Sí)')
plt.ylabel('Promedio de Ventas')
plt.grid(True)
plt.show();
```

```python
# Gráfico de barras para promedio de ventas por día de la semana
ventas_por_dia = df.groupby('DíaDeLaSemana')['Ventas'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=ventas_por_dia.index, y=ventas_por_dia.values, palette='plasma')
plt.title('Promedio de Ventas por Día de la Semana')
plt.xlabel('Día de la Semana (1 = Lunes, ..., 7 = Domingo)')
plt.ylabel('Promedio de Ventas')
plt.grid(True)
plt.show()
```

```python
modelo = linear_model.LinearRegression()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Extraer características temporales
df['Año'] = df['Fecha'].dt.year
df['Mes'] = df['Fecha'].dt.month
df['Día'] = df['Fecha'].dt.day
df['DíaDeLaSemana'] = df['Fecha'].dt.weekday  # 0 = Lunes, 6 = Domingo
df['EsFinDeSemana'] = df['DíaDeLaSemana'].apply(lambda x: 1 if x >= 5 else 0)

# Ordenar los datos por fecha
df = df.sort_values(by='Fecha')
```

```python
train = df[df['Fecha'] < '2022-12-01']  # Cambia la fecha según tu conjunto de datos
test = df[df['Fecha'] >= '2022-12-01']

# Separar variables independientes y dependientes
X_train = train.drop(columns=['Ventas', 'Fecha'])
y_train = train['Ventas']
X_test = test.drop(columns=['Ventas', 'Fecha'])
y_test = test['Ventas']
```

```python
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='Ventas Reales', color='blue', marker='o')
plt.plot(y_pred, label='Ventas Predichas', color='red', linestyle='--', marker='x')
plt.title('Ventas Reales vs. Ventas Predichas')
plt.xlabel('Índice')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)
plt.show()
```
