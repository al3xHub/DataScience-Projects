```python
import seaborn as sns
```

```python
datos = sns.load_dataset("diamonds")
```

```python
datos.head()
```

```python
datos.info()
```

```python
grafico = sns.relplot(data=datos, x="carat", y="price");
```

```python
grafico = sns.displot(data=datos, x="price")
grafico.fig.suptitle("Distribución del precio de los diamantes")
grafico.set_axis_labels('Precio ($)', 'Frecuencia');
```

```python
sns.displot(data=datos, x="carat");
```

```python
g = sns.jointplot(data=datos, x="depth", y="price", kind="kde", fill=True, color="red")
```

```python
g = sns.pairplot(datos, corner=True)
g.fig.suptitle("Relaciones entre variables numericas en el dataset Diamonds.", va="baseline", ha="center", fontsize=16)
```
